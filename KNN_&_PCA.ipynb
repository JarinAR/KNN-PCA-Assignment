**1.What is K-Nearest Neighbors (KNN) and how does it work?**

Ans: **K-Nearest Neighbors (KNN)** is a simple, **supervised machine learning algorithm** used for **classification and regression** tasks.

**How it works:**

1. **Choose a value of K** (number of neighbors to consider).
2. **Calculate the distance** (e.g., Euclidean) between the new data point and all existing data points.
3. **Select the K nearest neighbors** (based on distance).
4. **Classification:**
   - Assign the class **most common among the K neighbors**.
   **Regression:**
   - Take the **average (or weighted average)** of the K neighbors’ values.
5. **Prediction is made** based on majority vote (classification) or average (regression).

**Key points:**
- **Lazy learner** (no training phase).
- **Sensitive to outliers and irrelevant features.**
- **Works well with small datasets.**
- **K value selection is crucial** (too small = noise; too big = bias).

**2.What is the difference between KNN Classification and KNN Regression?**

Ans:**Difference between KNN Classification and KNN Regression:**

1. **Output Type**  
   - **Classification**: Predicts a **class label**  
   - **Regression**: Predicts a **numerical value**

2. **Decision Rule**  
   - **Classification**: Uses **majority voting**  
   - **Regression**: Uses **average** (or weighted average) of values

3. **Use Case Example**  
   - **Classification**: Spam or Not Spam  
   - **Regression**: Predict house price

4. **Evaluation Metric**  
   - **Classification**: Accuracy, Precision, Recall  
   - **Regression**: RMSE, MAE, R² Score

5. **Result Nature**  
   - **Classification**: Discrete output  
   - **Regression**: Continuous output

**3.What is the role of the distance metric in KNN?**

Ans: **Role of Distance Metric in KNN:**

1. **Measures similarity** between data points  
2. **Determines nearest neighbors** based on smallest distance  
3. **Affects prediction accuracy**—better distance = better neighbors  
4. **Common metrics used:**  
   - **Euclidean distance** (most common)  
   - **Manhattan distance**  
   - **Minkowski distance**  
   - **Cosine similarity** (for text/data in high dimensions)  
5. **Choice depends on data type and scale**  
   - Normalize/scale features to avoid bias from large values

**4.What is the Curse of Dimensionality in KNN?**

Ans: **Curse of Dimensionality in KNN:**

1. **Happens when data has too many features (dimensions)**  
2. **Distances between points become less meaningful**  
3. **All points appear almost equally distant**  
4. **KNN struggles to find true nearest neighbors**  
5. **Leads to poor accuracy and overfitting**

**Solution:**  
- Use **feature selection** or **dimensionality reduction** (like PCA)  
- **Normalize** the data before applying KNN

**5.How can we choose the best value of K in KNN?**

Ans: **Choosing the Best Value of K in KNN:**

1. **Try different K values** using **cross-validation**  
2. **Plot error vs. K** and select the K with **lowest error**  
3. **Odd K** is preferred in binary classification to avoid ties  
4. **Small K** (e.g., K=1) → low bias, high variance (overfitting)  
5. **Large K** → high bias, low variance (underfitting)  
6. Use techniques like **Grid Search** or **Elbow Method** to tune K

**Tip:** Start testing K values from 1 to √(number of samples)

**6.What are KD Tree and Ball Tree in KNN?**

Ans: **KD Tree and Ball Tree in KNN:**  
They are **data structures** used to **speed up nearest neighbor searches**.

1. **KD Tree (K-Dimensional Tree):**  
   - Binary tree for low-dimensional data  
   - Splits data by axis-aligned hyperplanes  
   - Efficient for **dimensions < 20**

2. **Ball Tree:**  
   - Groups data into hyperspheres (balls)  
   - Works better for **high-dimensional data**  
   - More flexible than KD Tree

3. **Why used:**  
   - Reduce time complexity of KNN search  
   - Avoid brute-force distance calculation for all points


**7.When should you use KD Tree vs. Ball Tree?**

Ans: **When to Use KD Tree vs. Ball Tree:**

1. **KD Tree:**
   - Best for **low-dimensional data** (≤ 20 features)
   - Faster for axis-aligned, structured data
   - More efficient with **balanced and uniform** data

2. **Ball Tree:**
   - Better for **high-dimensional data** (> 20 features)
   - Handles **non-uniform or clustered** data well
   - More flexible with **irregular data distributions**

**Rule of Thumb:**
- Use **KD Tree** for low dimensions  
- Use **Ball Tree** for high or complex dimensions

**8.What are the disadvantages of KNN?**

Ans: **Disadvantages of KNN:**

1. **Slow with large datasets** (no training phase, needs full scan)  
2. **High memory usage** (stores all training data)  
3. **Sensitive to irrelevant or noisy features**  
4. **Struggles with high-dimensional data** (curse of dimensionality)  
5. **No model interpretability** (black-box behavior)  
6. **Needs feature scaling** (distance-based algorithm)  
7. **Choice of K and distance metric is critical**  
8. **Can be biased by class imbalance**

**9.How does feature scaling affect KNN?**

Ans: **Effect of Feature Scaling on KNN:**

1. **KNN uses distance** (e.g., Euclidean), so **scale matters**  
2. **Unscaled features with larger ranges dominate the distance**  
3. Leads to **biased nearest neighbor selection**  
4. **Improper scaling → poor accuracy**

**Solution:**  
- Apply **feature scaling** (e.g., Min-Max Scaling or Standardization)  
- Ensures **all features contribute equally** to distance calculation

**10.What is PCA (Principal Component Analysis)?**

Ans: **PCA (Principal Component Analysis):**

- **Dimensionality reduction technique** used to reduce the number of features while preserving as much variance (information) as possible.
- **Transforms data** into a new coordinate system where:
  - The first axis (principal component) captures the most variance.
  - The second axis captures the next highest variance, and so on.
- **Goal**: Simplify data by projecting it onto fewer dimensions while retaining key patterns.

**How PCA works:**

1. **Standardize the data** (if needed).
2. **Compute the covariance matrix** of the data.
3. **Compute eigenvalues and eigenvectors** of the covariance matrix.
4. Sort eigenvectors by eigenvalue size, select the top N eigenvectors.
5. **Transform data** using selected eigenvectors (principal components).

**Benefits:**
- Reduces computational cost.
- Helps in visualizing high-dimensional data.
- Removes correlated features.

**11.How does PCA work?**

Ans: **How PCA Works:**

1. **Standardize the Data:**  
   - If the features have different scales, standardize them (mean = 0, variance = 1).

2. **Calculate the Covariance Matrix:**  
   - Measures the relationship (correlation) between different features.

3. **Compute Eigenvalues and Eigenvectors:**  
   - Eigenvectors represent the **directions** (principal components) in the data.
   - Eigenvalues indicate the **variance** (amount of information) in each direction.

4. **Sort Eigenvectors:**  
   - Rank eigenvectors by their eigenvalues in descending order, selecting the top ones that capture the most variance.

5. **Project the Data onto New Axes:**  
   - Transform the original data by projecting it onto the selected principal components, reducing dimensionality while preserving most information.

**12.What is the geometric intuition behind PCA?**

Ans: **Geometric Intuition behind PCA:**

1. **Data as points** in a multi-dimensional space.
2. **Find the direction of maximum variance** (principal component).
3. **Project data onto new axes** (principal components), which are orthogonal.
4. **Rotate the data** to align with the directions of maximum spread.
5. **Reduce dimensions** by removing axes with less variance, preserving key information.

PCA simplifies data by capturing the most important features (variance) while reducing dimensionality.

**13.What is the difference between Feature Selection and Feature Extraction?**

Ans: **Difference between Feature Selection and Feature Extraction:**

1. **Definition:**  
   - **Feature Selection:** Selects a subset of the original features, keeping the most relevant ones.
   - **Feature Extraction:** Combines existing features to create new, reduced features.

2. **Goal:**  
   - **Feature Selection:** Remove irrelevant or redundant features.
   - **Feature Extraction:** Reduce dimensionality by creating new features that summarize existing ones.

3. **Method:**  
   - **Feature Selection:** Chooses the best features based on statistical tests or algorithms (e.g., mutual information, chi-square).
   - **Feature Extraction:** Uses techniques like PCA, LDA to create new composite features.

4. **Result:**  
   - **Feature Selection:** Keeps original features.
   - **Feature Extraction:** Produces new features (e.g., principal components).

5. **Use Case:**  
   - **Feature Selection:** When features are already meaningful but redundant or irrelevant.
   - **Feature Extraction:** When data has too many features, and we want to transform them into a smaller, informative set.

**14.What are Eigenvalues and Eigenvectors in PCA?**

Ans: **Eigenvalues and Eigenvectors in PCA:**

1. **Eigenvectors:**
   - Represent the **directions** of maximum variance in the data.
   - In PCA, they define the **principal components** (new axes) onto which the data is projected.
   - Each eigenvector corresponds to one principal component.

2. **Eigenvalues:**
   - Measure the **magnitude of variance** along the corresponding eigenvector (principal component).
   - Larger eigenvalues indicate that the corresponding principal component explains more variance in the data.
   - The sum of all eigenvalues represents the total variance in the dataset.

**Role in PCA:**
- Eigenvectors determine the **directions** of maximum variance.
- Eigenvalues indicate the **importance** of those directions (i.e., how much variance each principal component captures).
  
PCA selects the eigenvectors with the largest eigenvalues to reduce the dimensionality while retaining most of the data's variability.

**15.How do you decide the number of components to keep in PCA?**

Ans: **Deciding the Number of Components in PCA:**

1. **Cumulative Explained Variance:**
   - Choose components that explain a desired amount of total variance (e.g., 90%-95%).

2. **Scree Plot:**
   - Look for the "elbow" where the explained variance starts to level off.

3. **Domain Knowledge:**
   - Consider problem context and how many components are needed for tasks.

4. **Cross-validation:**
   - Test different component counts to optimize model performance.

These methods help balance dimensionality reduction and data retention.

**16.Can PCA be used for classification?**

Ans: Yes, **PCA can be used for classification**, but indirectly. Here's how:

1. **Dimensionality Reduction:**  
   - PCA reduces the number of features (dimensions) while preserving most of the variance in the data.
   - This helps improve the performance of classification models, especially with high-dimensional data (e.g., image or text data).

2. **Preprocessing Step:**  
   - PCA can be applied **before** using a classification algorithm (e.g., SVM, Logistic Regression) to reduce overfitting and computational cost.

3. **Classifying on Principal Components:**  
   - After performing PCA, you can use the new **principal components** as input features for classification models.
   - This reduces the complexity and may enhance model performance, especially when original features are noisy or highly correlated.



**17.What are the limitations of PCA?**

Ans: **Limitations of PCA:**

1. **Linear Assumption:**
   - PCA assumes linear relationships between features, which may not capture complex, non-linear patterns in the data.

2. **Interpretability:**
   - The new features (principal components) are combinations of original features, making them hard to interpret.

3. **Sensitive to Scaling:**
   - PCA is sensitive to the **scale of features**, requiring proper **feature scaling** (e.g., standardization or normalization) before applying.

4. **Loss of Information:**
   - Reducing dimensions can result in **loss of important information** if too many components are discarded.

5. **Assumes Normality:**
   - PCA performs best when data is approximately **normally distributed**; it might not work well with skewed or non-Gaussian data.

6. **Outlier Sensitivity:**
   - PCA can be heavily influenced by **outliers**, as they can skew the principal components.


**18.How do KNN and PCA complement each other?**

Ans: **How KNN and PCA Complement Each Other:**

1. **Dimensionality Reduction for KNN:**  
   - PCA reduces the number of features in the dataset, making the KNN algorithm faster by decreasing the computational complexity.

2. **Improving KNN Performance:**  
   - PCA helps in removing **noise and correlated features**, improving the accuracy of KNN by focusing on the most important patterns in the data.

3. **Handling Curse of Dimensionality:**  
   - PCA addresses the **curse of dimensionality** in KNN, where high-dimensional spaces lead to less meaningful distance calculations. By reducing dimensions, PCA helps KNN find more accurate nearest neighbors.

4. **Enhanced Visualization:**  
   - PCA reduces dimensions to 2 or 3, making it easier to **visualize data** and understand how KNN is classifying the data, especially for exploratory analysis.

**19.How does KNN handle missing values in a dataset?**

Ans:**Handling Missing Values in KNN:**

1. **Imputation:**
   - **Mean/Median/Mode**: Replace missing values with mean (continuous), median, or mode (categorical).
   - **KNN Imputation**: Use nearest neighbors to impute missing values.

2. **Remove Missing Data:**
   - Delete rows/columns with missing values (may cause data loss).

3. **Distance Calculation Adjustment:**
   - Some implementations allow ignoring missing values during distance calculations.

Imputation is the most common and preferred approach for handling missing values.

**20.What are the key differences between PCA and Linear Discriminant Analysis (LDA)?**

Ans:**Key Differences between PCA and LDA:**

1. **Objective:**
   - **PCA:** Maximizes variance in the data (unsupervised).
   - **LDA:** Maximizes class separability (supervised).

2. **Data Type:**
   - **PCA:** Works with any data (no class labels).
   - **LDA:** Requires labeled data (class labels).

3. **Dimensionality Reduction:**
   - **PCA:** Reduces dimensions by preserving variance.
   - **LDA:** Reduces dimensions by maximizing class separation.

4. **Feature Transformation:**
   - **PCA:** Creates principal components unrelated to classes.
   - **LDA:** Creates discriminant components related to class separation.

5. **Number of Components:**
   - **PCA:** At most the number of features.
   - **LDA:** At most **C - 1** components (C = number of classes).

PCA is for variance, LDA is for classification.

**21.Train a KNN Classifier on the Iris dataset and print model accuracy.**

```python
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Predict on the test set
y_pred = knn.predict(X_test)

# Calculate and print the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)
```

**24.Train a KNN Classifier with different values of K and visualize decision boundaries.**

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Load Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Use only the first two features for visualization
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Function to plot decision boundaries
def plot_decision_boundaries(X, y, model, title="Decision Boundary"):
    h = .02  # Step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o', s=50, cmap=plt.cm.Paired)
    plt.title(title)
    plt.show()

# Train and visualize KNN classifiers with different values of K
for k in [1, 3, 5, 7]:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    plot_decision_boundaries(X, y, knn, title=f"Decision Boundary for K={k}")
```

**25.Apply Feature Scaling before training a KNN model and compare results with unscaled data.**

```python
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize KNN with unscaled data
knn_unscaled = KNeighborsClassifier(n_neighbors=3)
knn_unscaled.fit(X_train, y_train)
y_pred_unscaled = knn_unscaled.predict(X_test)
accuracy_unscaled = accuracy_score(y_test, y_pred_unscaled)

# Apply feature scaling using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize KNN with scaled data
knn_scaled = KNeighborsClassifier(n_neighbors=3)
knn_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = knn_scaled.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)

# Print the accuracies for both unscaled and scaled data
print(f"Accuracy with unscaled data: {accuracy_unscaled}")
print(f"Accuracy with scaled data: {accuracy_scaled}")
```

**27.Apply PCA before training a KNN Classifier and compare accuracy with and without PCA.**

```python
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the KNN classifier without PCA
knn_no_pca = KNeighborsClassifier(n_neighbors=3)
knn_no_pca.fit(X_train, y_train)
y_pred_no_pca = knn_no_pca.predict(X_test)
accuracy_no_pca = accuracy_score(y_test, y_pred_no_pca)

# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 components for visualization purposes
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Initialize and train the KNN classifier with PCA
knn_with_pca = KNeighborsClassifier(n_neighbors=3)
knn_with_pca.fit(X_train_pca, y_train)
y_pred_with_pca = knn_with_pca.predict(X_test_pca)
accuracy_with_pca = accuracy_score(y_test, y_pred_with_pca)

# Print the accuracies with and without PCA
print(f"Accuracy without PCA: {accuracy_no_pca}")
print(f"Accuracy with PCA: {accuracy_with_pca}")
```

**28.Perform Hyperparameter Tuning on a KNN Classifier using GridSearchCV.**

```python
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the KNN classifier
knn = KNeighborsClassifier()

# Define parameter grid for hyperparameter tuning
param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

# Initialize GridSearchCV with 5-fold cross-validation
grid_search = GridSearchCV(knn, param_grid, cv=5)

# Fit GridSearchCV to the training data
grid_search.fit(X_train, y_train)

# Get the best parameters from the grid search
best_params = grid_search.best_params_

# Train the KNN classifier with the best parameters
best_knn = grid_search.best_estimator_

# Evaluate the model
accuracy = best_knn.score(X_test, y_test)

# Print the best parameters and the accuracy
print(f"Best Hyperparameters: {best_params}")
print(f"Accuracy with best hyperparameters: {accuracy}")
```

**31.Train a KNN Classifier using different values of the weights parameter (uniform vs. distance) and compare accuracy.**

**38.Train a PCA model and analyze the effect of different numbers of components on accuracy.**

```python
# Import necessary libraries
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# List of different numbers of components to test
n_components_list = [1, 2, 3, 4]

# Store accuracies for each number of components
accuracies = []

# Train and evaluate KNN classifier with different numbers of PCA components
for n_components in n_components_list:
    # Apply PCA
    pca = PCA(n_components=n_components)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)

    # Initialize and train the KNN classifier
    knn = KNeighborsClassifier(n_neighbors=3)
    knn.fit(X_train_pca, y_train)

    # Predict and calculate accuracy
    y_pred = knn.predict(X_test_pca)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

# Plot the effect of different numbers of components on accuracy
plt.plot(n_components_list, accuracies, marker='o')
plt.title('Effect of Number of PCA Components on Accuracy')
plt.xlabel('Number of Components')
plt.ylabel('Accuracy')
plt.show()

# Print the accuracies for each number of components
for n_components, accuracy in zip(n_components_list, accuracies):
    print(f"Accuracy with {n_components} components: {accuracy}")
```

**42.Train a KNN Regressor and analyze the effect of different distance metrics on prediction error.**

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Create synthetic regression data
X, y = make_regression(n_samples=100, n_features=1, noise=0.1, random_state=42)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# List of different distance metrics to test
distance_metrics = ['euclidean', 'manhattan', 'chebyshev', 'minkowski']

# Store MSE for each distance metric
mse_values = []

# Train and evaluate KNN regressor for each distance metric
for metric in distance_metrics:
    # Initialize and train the KNN regressor with the current distance metric
    knn_regressor = KNeighborsRegressor(n_neighbors=3, metric=metric)
    knn_regressor.fit(X_train, y_train)

    # Predict and calculate Mean Squared Error (MSE)
    y_pred = knn_regressor.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_values.append(mse)

# Plot the effect of different distance metrics on MSE
plt.bar(distance_metrics, mse_values)
plt.title('Effect of Different Distance Metrics on Prediction Error')
plt.xlabel('Distance Metric')
plt.ylabel('Mean Squared Error (MSE)')
plt.show()

# Print MSE for each distance metric
for metric, mse in zip(distance_metrics, mse_values):
    print(f"MSE with {metric} distance: {mse}")
```

**43.Train a KNN Classifier and evaluate using ROC-AUC score.**

```python
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import label_binarize

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Binarize the labels for multi-class ROC-AUC
y_bin = label_binarize(y, classes=[0, 1, 2])

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_bin, test_size=0.3, random_state=42)

# Initialize and train the KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Predict probabilities
y_pred_prob = knn.predict_proba(X_test)

# Since the classifier is multi-class, flatten the probabilities to 2D
y_pred_prob_flat = np.array([y_pred_prob[i][:, 1] for i in range(len(y_pred_prob))]).T

# Compute ROC-AUC score for each class
roc_auc = roc_auc_score(y_test, y_pred_prob_flat, average='macro', multi_class='ovr')

# Print the ROC-AUC score
print(f"ROC-AUC score: {roc_auc}")
```

**46.Train a PCA model and visualize the data reconstruction error after reducing dimensions.**

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Load Iris dataset
iris = load_iris()
X = iris.data

# Split the dataset into training and testing sets
X_train, X_test = train_test_split(X, test_size=0.3, random_state=42)

# Initialize PCA model
pca = PCA(n_components=2)

# Fit PCA on the training data and transform both train and test data
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Reconstruct the data from the reduced dimensions
X_train_reconstructed = pca.inverse_transform(X_train_pca)
X_test_reconstructed = pca.inverse_transform(X_test_pca)

# Calculate the reconstruction error (Mean Squared Error)
train_error = np.mean((X_train - X_train_reconstructed) ** 2)
test_error = np.mean((X_test - X_test_reconstructed) ** 2)

# Print the reconstruction errors
print(f"Training Data Reconstruction Error: {train_error}")
print(f"Test Data Reconstruction Error: {test_error}")

# Plot the data reconstruction error
labels = ['Train Data', 'Test Data']
errors = [train_error, test_error]

plt.bar(labels, errors)
plt.title('Data Reconstruction Error After PCA')
plt.ylabel('Mean Squared Error')
plt.show()
```

**47. Train a KNN Classifier and visualize the decision boundary.**

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Load Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Use only the first two features for 2D visualization
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Create a meshgrid for plotting decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

# Predict class for each point in the meshgrid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o', s=50)
plt.title('KNN Classifier Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
